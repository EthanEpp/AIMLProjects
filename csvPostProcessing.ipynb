```python
import pandas as pd
```

```python

# Step 2: Read the CSV file
file_path = '/content/drive/MyDrive/school/capstone/urlAndLocationAndCoordsV5.csv'
df = pd.read_csv(file_path)

# Manual coordinates for specified locations
manual_coords = {
    'Alabama': '(32.806671, -86.791130)',
    'Oregon': '(44.572021, -122.070938)',
    'Los Angeles': '(34.052234, -118.243685)',
    'Santa Barbara': '(34.420831, -119.698190)'
}

# Update DataFrame with manual coordinates for specified locations, avoiding overwriting NaN values
for location, coords in manual_coords.items():
    # Identify rows where 'Locations' matches and 'Coordinates' is NaN (not set)
    condition = (df['Locations'] == location) & (df['Coordinates'].isna())
    # Update only those rows with the specified coordinates
    df.loc[condition, 'Coordinates'] = coords

# Delete any rows that do not have coordinates
df = df.dropna(subset=['Coordinates'])

# Define the output file path
output_file_path = '/content/drive/MyDrive/school/capstone/postProcessedNoEmptyCoords.csv'

# Write the filtered DataFrame to a new CSV file
df.to_csv(output_file_path, index=False)

print(f"DataFrame successfully written to {output_file_path}")
```

```python

# Step 2: Read the CSV file
file_path = '/content/drive/MyDrive/school/capstone/urlAndLocationAndCoordsOld.csv'
df = pd.read_csv(file_path)


# Ensure 'Coordinates' column is treated as string
df['Coordinates'] = df['Coordinates'].astype(str)

# Function to check if a coordinate is in North America
def is_in_north_america(coords):
    try:
        # Strip parentheses and split by comma, then convert to float
        lat, lon = map(float, coords.strip('()').split(','))
        return 7 <= lat <= 83 and -168 <= lon <= -52
    except ValueError:
        # Returns False if there's an error converting coordinates
        return False

# Apply the function to each row in the 'Coordinates' column and sum the True values
df['in_north_america'] = df['Coordinates'].apply(is_in_north_america)
north_america_count = df['in_north_america'].sum()

print(f"Number of coordinates within North America: {north_america_count}")
```

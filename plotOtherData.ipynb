```python
import matplotlib.pyplot as plt

import math
```

```python
dataEp = [[1,4/30],[2,6/30],[3,8/30],[4,2/30],[5,10/30]]


x, y = zip(*dataEp)  # unpack the data into two lists

plt.scatter(x, y)  # use scatter plot to show just the data points
plt.xticks(range(1, 6))  # set x-axis tick labels to integers only
plt.xlabel('Class Number')
plt.ylabel('Estimated Probability')
plt.title('Relative Frequency')
plt.show()  # display the plot
```
